```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/notyourfriends/Kaggle/main/Datasets/cancer.csv'
cancer_data = pd.read_csv(url, encoding = 'unicode_escape')
cancer_data
```

#Using Line Charts

```python
plt.figure(figsize=(16, 6))
sns.lineplot(data=cancer_data['Radius (mean)'])
plt.title('Radius (mean)')
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=cancer_data['Radius (mean)'], y=cancer_data['Radius (worst)'])
```

# Scatter plot

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x=cancer_data['Area (mean)'], y=cancer_data['Area (worst)'])
```

```python
sns.scatterplot(x=cancer_data['Area (mean)'], y=cancer_data['Area (worst)'], hue=cancer_data['Diagnosis'])
```

using sns.regplot  to draw line

```python
plt.figure(figsize=(15,5))
sns.regplot(x=cancer_data['Area (mean)'], y=cancer_data['Area (worst)'])
```

using sns.lmplot to draw line with color

```python
sns.lmplot(x='Area (mean)', y='Area (worst)', hue='Diagnosis', data=cancer_data)
```

using swarmplot

```python
sns.swarmplot(x=cancer_data['Diagnosis'], y=cancer_data['Area (mean)'])
```

# Distribution

using Histogram

```python
sns.histplot(data=cancer_data, x='Area (mean)', hue='Diagnosis')
```

using KDE (kernel density estimate)

```python
sns.kdeplot(data=cancer_data, x='Radius (worst)', hue='Diagnosis', shade=True) 
```

#Missing Data

```python
import numpy as np

np.random.seed(0)
```

```python
#get the number of missing data points per column
missing_values_count = cancer_data.isnull().sum()

#how many total missing values do we have?
total_cells = np.product(cancer_data.shape)
total_missing = missing_values_count.sum()

#percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
```
